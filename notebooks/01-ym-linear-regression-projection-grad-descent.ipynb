```python
import jax
import jax.numpy as jnp

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import arviz as az

import numpyro
import numpyro.distributions as dist
from numpyro.infer import MCMC, NUTS
```

```python
DATASET_URL = "https://raw.githubusercontent.com/rmcelreath/rethinking/master/data/WaffleDivorce.csv"
df = pd.read_csv(DATASET_URL, sep=";")
df.head()
```

```python
def model(X, y=None):
    theta0 = numpyro.sample("theta0", dist.Normal(0.0, 1.0)).reshape(-1, 1)
    theta1 = numpyro.sample("theta1", dist.Normal(0.0, 1.0)).reshape(-1, 1)
    Theta = jnp.concatenate((theta0, theta1), axis=-1)
    mu = jnp.matmul(X, Theta.T).reshape(-1)
    sigma = numpyro.sample("sigma", dist.Exponential(1.0))
    numpyro.sample("obs", dist.Normal(mu, sigma), obs=y)
```

```python
X = df[["Marriage", "MedianAgeMarriage"]].to_numpy()
y = df["Divorce"].to_numpy()
print(X.shape, y.shape)
```

```python
# Start from this source of randomness. We will split keys for subsequent operations.
rng_key = jax.random.PRNGKey(0)
rng_key, rng_key_ = jax.random.split(rng_key)

# Run NUTS
kernel = NUTS(model)
mcmc = MCMC(kernel, num_warmup=1000, num_samples=1000, num_chains=4)
mcmc.run(rng_key_, X, y)
mcmc.print_summary()
```

```python
posterior = mcmc.get_samples()
theta0 = posterior["theta0"].reshape(-1, 1)
theta1 = posterior["theta1"].reshape(-1, 1)
theta = jnp.concatenate((theta0, theta1), axis=-1)
sigma = posterior["sigma"].reshape(-1, 1)
print(theta.shape, sigma.shape)
```

```python
def project_grad(
    theta, sigma, X, num_params=1, num_epochs=100, learning_rate=0.1, rng_seed=300416
):
    """Primary projection method for a Bayesian linear regression.
    
    The objective function for this method is taken from Equation (5) in
    @Piirnonen2016, and the projection of the linear model variance is from
    Equation (7). Note that an analytic solution to this specific problem is
    given in Equation (6), but we have chosen to implement a gradient descent
    projection for generality.
    
    To do:
        * Project sigma with gradient descent
    
    Args:
        theta (ndarray): MCMC samples of the learned full parameters
        sigma (ndarray): MCMC samples of the learned regression noise
        X (ndarray): Full data space
        num_params (int): The size of the restricted parameter space
        num_epochs (int): The number of epochs for which to run Newton's 
            Method.
        learning_rate (float): Gradient descent learning rate
        rng_seed (int): Random number generator seed.

    Returns:
        ndarray: Restricted rojection of the parameter parameters
    """

    def _obj_fun(theta_perp, theta):
        """Kullback-Leibler distance multivariate objective function.
        
        Args:
            theta_perp (ndarray): The vector of variables to optimise.
            
        Returns:
            ndarray: The value of the objective function at the current
                values of `theta_perp`.
        """

        f = X @ theta
        f_perp = X_perp @ theta_perp
        return (f - f_perp).T @ (f - f_perp)
    
    def _optim_loop(theta_perp, theta):
        """Optimisation loop for gradient descent.
        
        This is separated from the main function for ease of vectorisation.
        """
        
        for epoch in range(num_epochs):
            theta_perp -= learning_rate * obj_grad(theta_perp, theta)
        return theta_perp

    # extract the number of observations and MCMC samples
    n = X.shape[0]
    m = theta.shape[0]
    # build restricted data space
    X_perp = X[:, :num_params]
    # initialise reasonable `theta_perp` to optimise
    theta_perp = np.zeros((m, num_params))
    obj_grad = jax.jit(jax.grad(_obj_fun, argnums=0))
    # run vectorised optimisation
    theta_perp = jax.vmap(_optim_loop, in_axes=(0, 0), out_axes=0)(theta_perp, theta)
    
    # TODO: project sigma with gradient descent
    sigma_perp = sigma
    
    return theta_perp, sigma_perp
```

```python
theta_perp, sigma_perp = project_grad(theta, sigma, X, learning_rate=10e-7)
print(theta_perp.shape, theta.shape, sigma_perp.shape, sigma.shape)
```

```python
datadict = {
    "theta_0_perp": theta_perp,
    "theta_0": theta0,
    "theta_1": theta1,
    "sigma_perp": sigma_perp,
    "sigma": sigma,
}
dataset = az.convert_to_inference_data(datadict)
dataset
```

```python
az.plot_posterior(dataset)
plt.show()
```

```python
def get_proj_dist(theta, theta_perp, X, X_perp):
    """Compute the distance from the projection to full parameters.
    
    Compute the Kullback-Leibler distance between the full
    parameters and the projected parameters given some full and 
    restricted data sets.
    """
    
    def _kl(theta, theta_perp):
        f = X @ theta
        f_perp = X_perp @ theta_perp
        return (f - f_perp).T @ (f - f_perp)

    return jax.vmap(_kl, in_axes=(0,0), out_axes=0)(theta, theta_perp)
```

```python
X_perp = X[:, :1]
proj_dists = get_proj_dist(theta, theta_perp, X, X_perp)

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
plt.hist(proj_dists, bins=30)
plt.title("Projection distances")
plt.xlabel("KL Distance")
plt.ylabel("Count")
plt.show()
```
